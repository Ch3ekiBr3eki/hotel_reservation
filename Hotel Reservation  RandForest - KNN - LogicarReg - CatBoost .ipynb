### Импортирую библиотеке для разделения на тестовые и тренинг даты + метрики + уберём предупреждения

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
warnings.filterwarnings("ignore", category=UserWarning) 
```

```python
X = df[['no_of_adults','no_of_children', 'no_of_weekend_nights', 'no_of_week_nights', 'type_of_meal_plan',	'required_car_parking_space', 'room_type_reserved',	'lead_time', 'arrival_month', 'market_segment_type', 'repeated_guest', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled', 'avg_price_per_room', 'no_of_special_requests']] 
y = df['booking_status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

### Начнём с простого - с логистической регрессии 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logical = LogisticRegression()
logical.fit(X_train,y_train)
```

```python
logical_predict = logical.predict(X_test)
```

### Сделаем табличку с результатами предсказаний для наглядности

```python
acc_lr = accuracy_score(y_test,logical_predict)
f1_lr = f1_score(y_test,logical_predict)
prec_lr = precision_score(y_test,logical_predict)
rec_lr = recall_score(y_test,logical_predict)
results_lr = pd.DataFrame([['Логистическая регрессия', acc_lr, f1_lr, prec_lr, rec_lr]],
                     columns =  ['Модель','Accuracy Score','F1 score','Precision','Recall'])
results_lr
```

##### Accuracy Score - показывает, как часто модель правильно предсказывает целевую переменную (в данном случае, успешно забронированный номер).  Метрика может быть полезна, если мы хотим оценить, насколько хорошо модель справляется с общим количеством предсказаний.

##### F1 Score - сбалансированная метрика, учитывающая как точность, так и полноту. Делает компромисс между точностью и полнотой, чтобы получить более объективную оценку производительности модели.

##### Precision - показывает, как много из предсказанных положительных результатов действительно являются положительными. Такая метрика может быть полезна, если мы хотим оценить, насколько точно модель предсказывает успешные бронирования.

##### Recall - показывает, как много из положительных результатов было правильно определено. Эта метрика может быть полезна, если мы хотим оценить, насколько хорошо модель находит успешные бронирования.
#### Условно говоря, если для нас важно минимизировать количество ложноположительных результатов (когда модель предсказывает успешное бронирование, но на самом деле оно не состоялось), то мы можем сосредоточиться на Precision. Если наша цель максимизировать количество успешных бронирований, то наиболее важным будет Recall.
